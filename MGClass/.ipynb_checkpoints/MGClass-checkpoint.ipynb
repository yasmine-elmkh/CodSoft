```python
import pandas as pd
import warnings          #importing warnings packege for ignoring warning messages 


warnings.filterwarnings("ignore")
data = pd.read_csv("Descriptions.txt")
data

#Create and apply fct to read data by splitting by ":::"
def LoadData(FilePath):
    with open(FilePath, 'r', encoding = 'utf-8') as file:
         data = file.readlines()
    data = [line.strip().split(':::') for line in data]
    return data
   
TrainData = LoadData("EntrainerData.txt")
TrainDf = pd.DataFrame(TrainData, columns = ['ID', 'Titre', 'Genre', 'Description'])   #Convert to dataframe and rename columns with proper columns names

TestData = LoadData("TestData.txt")
TestDf = pd.DataFrame(TestData, columns = ['ID', 'Titre', 'Description'])   # convert test data on dataframe and rename columns with proper columns names

TestSaolutionData = LoadData("TestDataSolution.txt")
TestSaolutionDf = pd.DataFrame(TestSaolutionData, columns = ['ID', 'Titre', 'Genre', 'Description'])   #Convert to dataframe and rename columns with proper columns names

#print("Train Data:")
#TrainDf 
#print("\nTest Data:")
#TestDf

#Feature extraction : TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer  

vectorizer = TfidfVectorizer(max_features = 20000)

XTrainFit = vectorizer.fit_transform(TrainDf["Description"]) #fit and transform the training datafram column named Description. / fit_transform():Le modèle apprend et transforme en même temps (souvent utilisé sur le train).
XTestFit = vectorizer.transform(TestDf["Description"])       #on fait pas le fit pour le testdata pour que le model ne triche pas! et alors accuracy grande! /transform():Le modèle utilise ce qu'il a appris pour transformer de nouvelles données. 

#print(f"Training data shape: {XTrainFit.shape}")   #display the shape of the training dataset
#print (f"Testing data shape: {XTestFit.shape}")

#encoding the target labels
from sklearn.preprocessing import LabelEncoder

LabelEncoder = LabelEncoder()      #encoding values into numeric labels

YTrain = LabelEncoder.fit_transform(TrainDf['Genre'])    # encoding or transforming dataframe column into numeric label

#print(f"Unique genres in the training data: {LabelEncoder.classes_}")   #check the uniqque genres after

#Model building with logistic regression
from sklearn.linear_model import LogisticRegression

LR_Model = LogisticRegression(max_iter = 200)   #building a logistic regression with maximum nbr of iterations of 200
LR_Model.fit(XTrainFit, YTrain)      #training he model with training dataset

YPred = LR_Model.predict(XTestFit)   #predict on the test dataset values
PredGenres = LabelEncoder.inverse_transform(YPred)   #apply inverse transform to predicted values

TestDf['Pred_Genre'] = PredGenres       #add predicted to the dataframe as a new column
TestDf[['Titre','Pred_Genre']]          #display the test dataframe with predictiond

TestDf['Pred_Genre'] = PredGenres

mergeDf = pd.merge(TestSaolutionDf[['ID', 'Genre']], TestDf[['ID', 'Pred_Genre']], on = 'ID')   #merge testsolutionDF with testDf
mergeDf

#Model evaluation Logistic regression
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(mergeDf['Genre'], mergeDf['Pred_Genre'])   # get eccuracy: compare actual and predict genres
print(f"Accuracy:{accuracy:.4f}")

print("\nclassification report:")                                     #to get classification report
print(classification_report(mergeDf['Genre'], mergeDf['Pred_Genre'])) #display and print classification report

```
